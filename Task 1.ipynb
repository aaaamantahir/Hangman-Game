```python
import random
import tkinter as tk
from tkinter import messagebox

# List of words to guess from
words_list = ['computer', 'laptop']

# Function to start the game
def start_game():
    global word, guessed, tries, max_tries
    word = random.choice(words_list)
    guessed = ['_'] * len(word)
    tries = 0
    max_tries = 6
    update_display()

# Function to handle the guess
def guess_letter():
    global tries
    letter = entry.get().lower()
    entry.delete(0, tk.END)
    
    if len(letter) != 1 or not letter.isalpha():
        messagebox.showinfo("Invalid Input", "Please enter a single letter.")
        return
    
    if letter in word:
        for i in range(len(word)):
            if word[i] == letter:
                guessed[i] = letter
    else:
        tries += 1
    
    update_display()
    
    if ''.join(guessed) == word:
        messagebox.showinfo("Hangman", "Congratulations! You've guessed the word!")
        start_game()
    elif tries >= max_tries:
        messagebox.showinfo("Hangman", f"Game Over! The word was '{word}'.")
        start_game()

# Function to update the display
def update_display():
    word_label.config(text=' '.join(guessed))
    tries_label.config(text=f"Tries: {tries}/{max_tries}")

# Initialize the main window
root = tk.Tk()
root.title("Hangman Game")

# Create and place widgets
word_label = tk.Label(root, font=('Helvetica', 18))
word_label.pack(pady=10)

tries_label = tk.Label(root, font=('Helvetica', 14))
tries_label.pack(pady=5)

entry = tk.Entry(root, font=('Helvetica', 14))
entry.pack(pady=10)
entry.bind("<Return>", lambda event: guess_letter())

guess_button = tk.Button(root, text="Guess", command=guess_letter, font=('Helvetica', 14))
guess_button.pack(pady=5)

start_button = tk.Button(root, text="Start New Game", command=start_game, font=('Helvetica', 14))
start_button.pack(pady=20)

# Start the first game
start_game()

# Run the main loop
root.mainloop()
```
